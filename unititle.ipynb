```python
import streamlit as st
from tensorflow.keras.models import load_model
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
st.title("UI Load & Use Model dari Google Drive atau Lokal")

st.write("""
Masukkan path lengkap file model Keras (.h5) yang sudah Anda simpan, lalu tekan tombol Load Model.
Model akan dimuat dan Anda bisa lakukan prediksi sederhana.
""")

model_path = st.text_input("/content/drive/MyDrive/skripsi/model_tomat.h5:", "")

model = None
if st.button("Load Model"):
    try:
        model = load_model(model_path)
        st.success("Model berhasil diload!")
        st.write(model.summary())
    except Exception as e:
        st.error(f"Gagal load model: {e}")

# Contoh prediksi sederhana bila model berhasil diload
if model is not None:
    st.subheader("Prediksi Sederhana")

    input_shape = model.input_shape[1:]  # Misal (224, 224, 3)
    st.write(f"Input shape model: {input_shape}")

    # Buat input dummy/random sesuai input shape model
    if st.button("Lakukan Prediksi pada Data Dummy Random"):
        dummy_input = np.random.random((1,) + input_shape)
        pred = model.predict(dummy_input)
        st.write(f"Hasil prediksi:\n{pred}")
```
